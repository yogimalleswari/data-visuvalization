```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Indian_Kids_Screen_Time.csv')
df.head()
```

```python
df.info()
df.columns
```

```python
# Create Age Bands
bins = [0, 5, 8, 12, 15, 18]
labels = ['Toddler (0-5)', 'Early Child (6-8)',
          'Pre-Teen (9-12)', 'Teen (13-15)', 'Late Teen (16-18)']

df['Age_Band'] = pd.cut(df['Age'], bins=bins, labels=labels)

df[['Age', 'Age_Band']].head()
```

```python
# The 'Week_Type' column does not exist in the DataFrame.
# Plotting 'Primary_Device' distribution as an alternative.
df['Primary_Device'].value_counts().plot(kind='bar')
plt.title("Distribution of Primary Devices")
plt.ylabel("Count")
plt.show()
```

```python
# The 'Week_Type' column does not exist in the DataFrame without a 'Date' column.
# Plotting 'Urban_or_Rural' distribution as an alternative.
df['Urban_or_Rural'].value_counts().plot(kind='bar')
plt.title("Urban vs Rural Usage")
plt.ylabel("Count")
plt.show()
```

```python
device_share = df['Primary_Device'].value_counts(normalize=True) * 100
print(device_share)
```

```python
device_share.plot(kind='pie', autopct='%1.1f%%')
plt.title("Device Usage Share")
plt.ylabel("")
plt.show()
```
