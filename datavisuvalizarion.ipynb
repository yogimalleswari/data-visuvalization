```python

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Indian_Kids_Screen_Time43.csv')

print("First 5 Rows:")
print(df.head())

print("\nColumn Names:")
print(df.columns)
```

```python
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

```python
for col in numerical_cols:
    plt.figure()
    plt.hist(df[col].dropna(), bins=10)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

    print(f"\nStatistical Summary of {col}")
    print(df[col].describe())
```

```python
for col in categorical_cols:
    counts = df[col].value_counts()

    plt.figure()
    plt.bar(counts.index, counts.values)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

    print(f"\nValue Counts of {col}")
    print(counts)
```

```python
plt.figure()
plt.hexbin(df['Age'], df['Avg_Daily_Screen_Time_hr'], gridsize=20)
plt.colorbar()
plt.title("Age vs Screen Time (Hexbin)")
plt.xlabel("Age")
plt.ylabel("Screen Time Hours")
plt.show()
```

```python
for num_col in numerical_cols:
    for cat_col in categorical_cols:
        plt.figure()
        df.boxplot(column=num_col, by=cat_col)
        plt.title(f"{num_col} by {cat_col}")
        plt.suptitle("")
        plt.xlabel(cat_col)
        plt.ylabel(num_col)
        plt.show()
```

```python
for i in range(len(categorical_cols)):
    for j in range(i+1, len(categorical_cols)):
        cross_tab = pd.crosstab(df[categorical_cols[i]], df[categorical_cols[j]])

        cross_tab.plot(kind='bar', stacked=True)
        plt.title(f"{categorical_cols[i]} vs {categorical_cols[j]}")
        plt.xlabel(categorical_cols[i])
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.show()

        print(f"\nCross Tab between {categorical_cols[i]} and {categorical_cols[j]}")
        print(cross_tab)
```
